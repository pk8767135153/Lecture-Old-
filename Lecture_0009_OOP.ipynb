# Fraction Class

## __ _str_ __  (self)
    When we won't print object then it will goes from __str__(self) and access the formate of object or access some message.
    __str__(self) is alwayes return something.

     if we write pass inside the  __str__(self) magicl method it will return NoneType and produce the error 

```python
class Fraction:
    def __init__(self):
        print("Non-Parameteried Constructor")
    # Define Parameteried Constructor
    def __init__(self,x,y):
        self.num = x
        self.den = y

#---------------------------------------------------------------------------------#
        
    def __str__(self):
        return f"{self.num}/{self.den} "

#---------------------------------------------------------------------------------#
    
    def __add__(self,other):
        print("__add__(self)")
        new_num = self.num*other.den + other.num*self.den
        new_den = self.den*other.den
        return "{}/{}".format(new_num,new_den)
    
    
#---------------------------------------------------------------------------------#
    def __sub__(self,other):
        print("__sub__(self,other)")
        new_num = self.num*other.den - other.num*self.den
        new_den = self.den*other.den
        return "{}/{}".format(new_num,new_den)

#----------------------------------------------------------------------------------#

    def __mul__(self,other):
        print("__mul__(self,other)")
        new_num = self.num*other.num 
        new_den = self.den*other.den
        return "{}/{}".format(new_num,new_den)
    
```

```python
fr1 = Fraction(1,2)
fr2 = Fraction(3,4)
print(fr1*fr2)
print(fr1+fr2)
print(fr1-fr2)
print(fr1)
```
