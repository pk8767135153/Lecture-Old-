## Aggregation : (Has A Relation) One class own other class

#### In Aggregation we cannot access privite variable in own class so we can use method for that aggregation

## Common Pratice is when we can make instace variable we can make as a privite

```python
#  Here Customer class own Address class 

class Customer:
    def __init__(self,name,gender,address):
        self.name = name
        self.gender = gender
        self.address = address
         
    
    def edit_profile(self,new_name,new_gender,new_city,new_pin,new_state):
        self.name = new_name
        self.gender = new_gender

        #  here we can call the method which is present inside the Address class 
        self.address.edit_address(new_city,new_pin,new_state)

    
    def print_address(self):
        #  Here we can call the method get_city() because Custome Class does not access direct privite variable __city
        print(self.name,self.address.get_city(),self.address.pin, self.address.state)

class Address:
    def __init__(self,city,pin,state):
        # If we can make the city as a privite variable so we cannot access city variable outside the class so we can write method for printing city

        self.__city = city
        self.pin = pin
        self.state = state

    def get_city(self):
           return self.__city
    
    #  If user won't to edit address so we can make function
    def edit_address(self,new_city,new_pin,new_state):
        self.__city = new_city
        self.pin = new_pin
        self.state = new_state

add1 = Address("Pune",122032,"Maharastra")

#  Here we can pass add1 Object inside the Customer class 
cust1 = Customer("Pravinsingh","Male",add1)
cust1.print_address()

cust1.edit_profile("Vabhavi","Female","Solapur",12312,"Maharastra")
cust1.print_address()
```

# Abstraction

## Single Abstraction Method

```python

# It IMP to import ABC which is Abstract Basic Class
from abc import ABC,abstractclassmethod


# When we wont to Inherate Abstract Class its More import to inside the child class we can write Security method
# This is Abstract Class which is Child Class of ABC or ABC stand for Abstrat Basic Class 
class Bankapp(ABC):

    #  Concuret Code
    def database(self):
        print("Connected")
    
    # This is Abstract Method mean in this method we cannot write any code. 
    @abstractclassmethod
    def security(self):
        pass
```

```python
# In MobileApp class we inherate the Bankapp Class there we write abstract method inside the MobileApp Class in this method we write Security Code

class MobileApp(Bankapp):
    def mobile_login(self):
        print("Login into mobile.")

    #  This is Security is abstract method 
    #  if we cannot write security/ abstract method inside MobileApp Class we cannot create object 
    def security(self):
        pass
```

```python
obj = MobileApp()
```

## Multiple Abstraction Method

```python

# It IMP to import ABC which is Abstract Basic Class
from abc import ABC,abstractclassmethod


# When we wont to Inherate Abstract Class its More import to inside the child class we can write Security method
# This is Abstract Class which is Child Class of ABC or ABC stand for Abstrat Basic Class 
class BankApp(ABC):

    #  Concuret Code
    def database(self):
        print("Connected")
    
    # This is Abstract Method mean in this method we cannot write any code. 
    @abstractclassmethod
    def security(self):
        pass

    @abstractclassmethod
    def display(self):
        pass
```

```python
# In MobileApp class we inherate the Bankapp Class there we write abstract method inside the MobileApp Class in this method we write Security Code

class MobileApp1(BankApp):
    def mobile_login(self):
        print("Login into mobile.")

    #  This is Security is abstract method 
    #  if we cannot write security/ abstract method inside MobileApp Class we cannot create object 
    def security(self):
        # Write Sequrity Code Here
        pass

    # This is also abstract method
    def display(self):
        # Write Sequrity Code Here
        pass
```

```python
obj2 = MobileApp1()
```
