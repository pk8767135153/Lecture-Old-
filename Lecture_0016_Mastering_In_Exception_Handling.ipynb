### How to Handle Error

#### Using try , except block we can handle error


```python
#   POINT 0 : General Exception Hanling Error 
# Note : we can write alway General Exception hadling block at end Position.
try:
    print(a)
     
except Exception as e:
    print(e)
```

```python
## POINT 1 : One code contain Multiple Error 
try:
    print(a)
    print("b"+3)
except:
    print("Sorry Some Error is Occured. ")

    """
        In above code there is two error is occured so we can print message that is Sorry Some Error but in above code there is two error so it's not good way to handle error
    """
```

```python
## POINT 2 : Know about Error 

#-------------------------------------------------------#
try:
    # print(a)
    print("b"+3)
except Exception as e:
    print(e)

""" 
    In above code produce output what is main problem in that code.
    In short it will descibe problem brifely. 
    But It cannot return error name.
"""
#----------------------------------------------------------#
try:
    # print(a)
    print("b"+3)
except Exception as e:
    print(e.with_traceback)
    
    # It cannot return error name.

```

#### try, except with else block
#### When we used else block we can know try will defindetaly execuit without causeing error so we can write else block which run by complete exectuing of try block.<br> If try block produce error so else block not execuit

```python
try:
    a = 3+"get_ipython"
except TypeError:
    print("Error Is Found.")
except Exception:
    print("Error Occured.")
else:
    print(a)
```

#### try, except ,else block with finally block

#### finally always run no matter error occure or not occure

```python
try:
    a = 3+"get_ipython"
except TypeError:
    print("Error Is Found.")
except Exception:
    print("Error Occured.")
else:
    print(a)
finally:
    print("It will always run. not matter error occure or not occure")
```

## Python also provide to create custum exception class which handle error
#### When we can write custum exception class that time we can perform some specific task / operation.


```python
class MyException(Exception):
    def __init__(self,message):
        print(message)

class Bank:
    def __init__(self,balance):
        self.balance = balance
    
    def withdraw(self,amount):
        if amount<0:
            raise MyException("Amount Cannot be Negative.")
        if self.balance<amount:
            raise MyException("Money is not Suffecient.")
        self.balance = self.balance - amount

obj = Bank(10000)
try:
    obj.withdraw(4000)
except MyException as e:
    pass
else:
    print(obj.balance)
```
