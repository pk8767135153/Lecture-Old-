## Nested Function 
Nested Function is basecilly there is two fuction one is outer function and another is inside the function that means function inside the another function.<br><br>

we cannot access diretaly inner function. we can access the inner function with the help of outer function.<br><br>

Inner function having scope only limited scope

```python
def outer_fun():
    print("This is my outer functin.")
    def inner_fun():
        print("This is my Inner Function.")
    
    inner_fun()

outer_fun()
```

#### Another Example of Nested Function


```python
def cal():
    def add(a,b):
        return a+b
    def sub(a,b):
        return a-b
    def mul(a,b):
        return a*b
    def div(a,b):
        return a/b
    return{
        "add":add,
        "sub":sub,
        "mul":mul,
        "div":div
    }
cal()["div"](2,5)
```

```python
def outer_function(x):
    def inner_function(y):
         
        return y * 2
    
    result = inner_function(x)
    return result
output = outer_function(3)
print(output)
```

```python
def inner_function(x):
    def inner_function(y):
         
        return y * 2
    
    result = inner_function(x)
    return result
output = inner_function(3)
print(output)
```

```python
# find squre of given number
lambda a : a**2
# In this fucntion a is input
```

```python
# Store lambda function inside the variable
s = lambda a: a**2
s(4)
```

```python
def opearation(a,b,oper):
    return oper(a,b)
def add(a,b):
    return a+b
def mul(a,b):
    return a*b
    
# opearation(2,3,add)

# We can writen this fucntion in one line using lambda fuction
print("\n ",
    opearation(5,9,lambda a,b : a + b),"\n ",
    opearation(5,9,lambda a,b : a - b),"\n ",
    opearation(5,9,lambda a,b : a * b)
     )
```

### Map()

```python
# map(fun,iterable)
list(filter(lambda x: x%2==0, [1, 2, 3, 4, 5]))
```

```python
from functools import reduce 
reduce(lambda x, y: x+y, [1, 2, 3, 4, 5])
```

```python
l = [1,2,3,4,5,6,7]
list(filter(lambda x: len(x) > 3, ['cat', 'dog','elephant', 'lion']))
```

```python
a = map(lambda x : x*x,l)
list(a)
```

```python
word = ["Pravin","vabhavi Pawar","Jay"]
list(map(lambda word : word.title(),word))
```

```python
# Function to check if a year is a leap year
def is_leap_year(year):
 if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
    return True
 else:
    return False
# Input: Year to check
year = int(input("Enter a year: "))
# Determine and print if the year is a leap year
if is_leap_year(year):
 print(f"{year} is a leap year.")
else:
 print(f"{year} is not a leap year.")
```

```python
# Function to determine the grade based on marks
def determine_grade(marks):
    if marks >= 90:
        return 'A'
    elif marks >= 80:
        return 'B'
    elif marks >= 70:
        return 'C'
    elif marks >= 60:
        return 'D'
    else:
        return 'F'
# Input: Student's marks
marks = float(input("Enter the student's marks: "))
# Determine and print the grade
grade = determine_grade(marks)
print(f"The grade for marks {marks} is: {grade}")
```
